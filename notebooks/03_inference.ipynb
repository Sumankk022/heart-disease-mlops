# 03 - Inference

Loads the saved pipeline and makes predictions - the same artifact the API serves.

```python
import sys, os
sys.path.append(os.path.abspath('..'))
import joblib, pandas as pd
from src.config import MODEL_PATH, FEATURES
model = joblib.load(MODEL_PATH)
print('loaded', MODEL_PATH)
```

## Predict on sample patients

```python
samples = pd.DataFrame([
    {'age':67,'sex':1,'cp':4,'trestbps':160,'chol':286,'fbs':0,'restecg':2,
     'thalach':108,'exang':1,'oldpeak':1.5,'slope':2,'ca':3,'thal':3},
    {'age':45,'sex':0,'cp':2,'trestbps':120,'chol':180,'fbs':0,'restecg':0,
     'thalach':175,'exang':0,'oldpeak':0.2,'slope':1,'ca':0,'thal':3},
])[FEATURES]
preds = model.predict(samples)
proba = model.predict_proba(samples)
out = samples.copy()
out['prediction'] = preds
out['label'] = ['Heart disease' if p else 'No heart disease' for p in preds]
out['confidence'] = [round(float(proba[i][p]), 4) for i, p in enumerate(preds)]
out[['age','sex','cp','prediction','label','confidence']]
```
